## Data Types
## தரவு வகைகள்

To analyze data, it is important to know what type of data we are dealing with.
> தரவை பகுப்பாய்வு செய்ய, நாங்கள் எந்த வகையான தரவைக் கையாளுகிறோம் என்பதை அறிந்து கொள்வது அவசியம்.

We can split the data types into three main categories:

    * Numerical
    * Categorical
    * Ordinal

> தரவு வகைகளை மூன்று முக்கிய வகைகளாகப் பிரிக்கலாம்:
>
>    * எண்ணியல்
>    * வகைப்பாடு
>    * சாதாரண

Numerical data are numbers, and can be split into two numerical categories:

  * Discrete Data
     - numbers that are limited to integers. Example: The number of cars passing by.
  * Continuous Data
      - numbers that are of infinite value. Example: The price of an item, or the size of an item

> எண் தரவு எண்கள் மற்றும் இரண்டு எண் வகைகளாகப் பிரிக்கலாம்:

>  * தனித்தனி தரவு
     - முழு எண்களுக்கு மட்டுப்படுத்தப்பட்ட எண்கள். எடுத்துக்காட்டு: கடந்து செல்லும் கார்களின் எண்ணிக்கை.
>  * தொடர்ச்சியான தரவு
      - எல்லையற்ற மதிப்புள்ள எண்கள். எடுத்துக்காட்டு: ஒரு பொருளின் விலை அல்லது பொருளின் அளவு

**Categorical** data are values that cannot be measured up against each other. Example: a color value, or any yes/no values.
> **வகையான** தரவு என்பது ஒன்றுக்கொன்று எதிராக அளவிட முடியாத மதிப்புகள். எடுத்துக்காட்டு: வண்ண மதிப்பு அல்லது ஆம்/இல்லை மதிப்புகள்.

**Ordinal** data are like categorical data, but can be measured up against each other. Example: school grades where A is better than B and so on.
> **ஆர்டினல்** தரவு வகைப்பட்ட தரவு போன்றது, ஆனால் ஒன்றுக்கொன்று எதிராக அளவிட முடியும். எடுத்துக்காட்டு: B ஐ விட A சிறந்ததாக இருக்கும் பள்ளி தரங்கள் மற்றும் பல.

By knowing the data type of your data source, you will be able to know what technique to use when analyzing them.
> உங்கள் தரவு மூலத்தின் தரவு வகையை அறிந்துகொள்வதன் மூலம், அவற்றை பகுப்பாய்வு செய்யும் போது என்ன நுட்பத்தைப் பயன்படுத்த வேண்டும் என்பதை நீங்கள் அறிந்து கொள்ள முடியும்.

You will learn more about statistics and analyzing data in the next chapters.

# Mean Median Mode
# சராசரி இடைநிலை பயன்முறை

What can we learn from looking at a group of numbers?
>எண்களின் குழுவைப் பார்ப்பதன் மூலம் நாம் என்ன கற்றுக்கொள்ளலாம்?

In Machine Learning (and in mathematics) there are often three values that interests us:

  * **Mean** - The average value
  * **Median** - The mid point value
  * **Mode** - The most common value
> இயந்திர கற்றலில் (மற்றும் கணிதத்தில்) பெரும்பாலும் மூன்று மதிப்புகள் நமக்கு ஆர்வமாக உள்ளன:

>  * **சராசரி** - சராசரி மதிப்பு
>  * ** சராசரி** - நடுப்புள்ளி மதிப்பு
>  * **முறை** - மிகவும் பொதுவான மதிப்பு

Example: We have registered the speed of 13 cars:

## `speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]`

What is the average, the middle, or the most common speed value?
> சராசரி, நடுத்தர அல்லது மிகவும் பொதுவான வேக மதிப்பு என்ன?

## Mean
## சராசரி
The mean value is the average value.
> சராசரி மதிப்பு என்பது சராசரி மதிப்பு.

To calculate the mean, find the sum of all values, and divide the sum by the number of values:
> சராசரியைக் கணக்கிட, அனைத்து மதிப்புகளின் கூட்டுத்தொகையைக் கண்டறிந்து, மதிப்புகளின் எண்ணிக்கையால் கூட்டுத்தொகையைப் வகுக்க:

## `(99+86+87+88+111+86+103+87+94+78+77+85+86) / 13 = 89.77`

The NumPy module has a method for this. Learn about the NumPy module in our NumPy Tutorial.
> NumPy தொகுதி இதற்கு ஒரு முறையைக் கொண்டுள்ளது. எங்கள் NumPy டுடோரியலில் NumPy தொகுதி பற்றி அறிக.

```python
# Use the NumPy mean() method to find the average speed:

import numpy

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = numpy.mean(speed)

print(x)
```

## Median
## இடைநிலை

The median value is the value in the middle, after you have sorted all the values:
>நீங்கள் எல்லா மதிப்புகளையும் வரிசைப்படுத்திய பிறகு, நடுவில் உள்ள மதிப்பே சராசரி மதிப்பு:

## `77, 78, 85, 86, 86, 86, 87, 87, 88, 94, 99, 103, 111`

```python
# Use the NumPy median() method to find the middle value:

import numpy

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = numpy.median(speed)

print(x)
```

If there are two numbers in the middle, divide the sum of those numbers by two.
> நடுவில் இரண்டு எண்கள் இருந்தால், அந்த எண்களின் கூட்டுத்தொகையை இரண்டால் வகுக்கவும்.

## `77, 78, 85, 86, 86, 86, 87, 87, 94, 98, 99, 103`

## `(86 + 87) / 2 = 86.5`

```python
import numpy

speed = [99,86,87,88,86,103,87,94,78,77,85,86]

x = numpy.median(speed)

print(x)
```

## Mode
## பயன்முறை
The Mode value is the value that appears the most number of times:
> பயன்முறை மதிப்பு என்பது அதிக முறை தோன்றும் மதிப்பு:

## `99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86 = 86`

The SciPy module has a method for this. Learn about the SciPy module in our SciPy Tutorial.
> SciPy தொகுதி இதற்கு ஒரு முறையைக் கொண்டுள்ளது. எங்கள் SciPy டுடோரியலில் SciPy தொகுதி பற்றி அறிக.

```python
# Use the SciPy mode() method to find the number that appears the most:

from scipy import stats

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = stats.mode(speed)

print(x)
```

# Standard Deviation
# நிலையான விலகல்

## What is Standard Deviation?
## நிலையான விலகல் என்றால் என்ன?

Standard deviation is a number that describes how spread out the values are.

> நிலையான விலகல் என்பது மதிப்புகள் எவ்வாறு பரவுகின்றன என்பதை விவரிக்கும் எண்ணாகும்.

A low standard deviation means that most of the numbers are close to the mean (average) value.

> குறைந்த நிலையான விலகல் என்பது பெரும்பாலான எண்கள் சராசரி (சராசரி) மதிப்புக்கு அருகில் இருப்பதைக் குறிக்கிறது.

A high standard deviation means that the values are spread out over a wider range.

> உயர் தர விலகல் என்பது மதிப்புகள் பரந்த அளவில் பரவியிருப்பதைக் குறிக்கிறது.

Example: This time we have registered the speed of 7 cars:

> எடுத்துக்காட்டு: இந்த முறை நாங்கள் 7 கார்களின் வேகத்தை பதிவு செய்துள்ளோம்:

## `speed = [86,87,88,86,87,85,86]`

The standard deviation is:
## `0.9`

> நிலையான விலகல்:
> ## `0.9`

Meaning that most of the values are within the range of 0.9 from the mean value, which is 86.4.

> பெரும்பாலான மதிப்புகள் சராசரி மதிப்பிலிருந்து 0.9 வரம்பிற்குள் உள்ளன, அதாவது 86.4.

The NumPy module has a method to calculate the standard deviation:

>NumPy தொகுதி நிலையான விலகலைக் கணக்கிடுவதற்கான ஒரு முறையைக் கொண்டுள்ளது:

```python
# Use the NumPy std() method to find the standard deviation:

import numpy

speed = [86,87,88,86,87,85,86]

x = numpy.std(speed)

print(x)
```

Let us do the same with a selection of numbers with a wider range:

> பரந்த வரம்பைக் கொண்ட எண்களின் தேர்விலும் இதைச் செய்வோம்:

## `speed = [32,111,138,28,59,77,97]`

The standard deviation is:
> நிலையான விலகல்:
> ## `37.85`

Meaning that most of the values are within the range of 37.85 from the mean value, which is 77.4.

> பெரும்பாலான மதிப்புகள் சராசரி மதிப்பிலிருந்து 37.85 வரம்பிற்குள் உள்ளன, அதாவது 77.4.

As you can see, a higher standard deviation indicates that the values are spread out over a wider range.

> நீங்கள் பார்க்க முடியும் என, உயர் தர விலகல் மதிப்புகள் பரந்த அளவில் பரவியிருப்பதைக் குறிக்கிறது.

```python
import numpy

speed = [32,111,138,28,59,77,97]

x = numpy.std(speed)

print(x)
```

## Variance
## மாறுபாடு
Variance is another number that indicates how spread out the values are.

> மாறுபாடு என்பது மதிப்புகள் எவ்வாறு பரவுகின்றன என்பதைக் குறிக்கும் மற்றொரு எண்.

In fact, if you take the square root of the variance, you get the standard deviation!

> உண்மையில், நீங்கள் மாறுபாட்டின் வர்க்க மூலத்தை எடுத்துக் கொண்டால், நீங்கள் நிலையான விலகலைப் பெறுவீர்கள்!

Or the other way around, if you multiply the standard deviation by itself, you get the variance!

> அல்லது வேறு வழியில், நீங்கள் நிலையான விலகலைப் பெருக்கினால், நீங்கள் மாறுபாட்டைப் பெறுவீர்கள்!

To calculate the variance you have to do as follows:

> மாறுபாட்டைக் கணக்கிட, நீங்கள் பின்வருவனவற்றைச் செய்ய வேண்டும்:



    1.   Find the mean:
    1. சராசரியைக் கண்டறியவும்:
  ## `(32+111+138+28+59+77+97) / 7 = 77.4`

    2. For each value: find the difference from the mean:
    2. ஒவ்வொரு மதிப்புக்கும்: சராசரியிலிருந்து வித்தியாசத்தைக் கண்டறியவும்:
  ## `32 - 77.4 = -45.4`
  ## `111 - 77.4 =  33.6`
  ## ` 138 - 77.4 =  60.6`
  ## `28 - 77.4 = -49.4`
  ## `59 - 77.4 = -18.4`
 ## `77 - 77.4 = - 0.4`
 ## `97 - 77.4 =  19.6`

    3. For each difference: find the square value:
    3. ஒவ்வொரு வித்தியாசத்திற்கும்: சதுர மதிப்பைக் கண்டறியவும்:

  ## `(-45.4)2 = 2061.16`
 ## `(33.6)2 = 1128.96`
 ## `(60.6)2 = 3672.36`
## `(-49.4)2 = 2440.36`
## `(-18.4)2 =  338.56`
## `(- 0.4)2 =    0.16`
 ## `(19.6)2 =  384.16`

    4. The variance is the average number of these squared differences:
    4. மாறுபாடு என்பது இந்த வர்க்க வேறுபாடுகளின் சராசரி எண்ணிக்கை:

  ## `(2061.16+1128.96+3672.36+2440.36+338.56+0.16+384.16) / 7 = 1432.2`

```python
# Use the NumPy var() method to find the variance:

import numpy

speed = [32,111,138,28,59,77,97]

x = numpy.var(speed)

print(x)
```

## Standard Deviation
## நிலையான விலகல்
As we have learned, the formula to find the standard deviation is the square root of the variance:

> நாம் கற்றுக்கொண்டபடி, நிலையான விலகலைக் கண்டுபிடிப்பதற்கான சூத்திரம் மாறுபாட்டின் வர்க்க மூலமாகும்:

## `√1432.25 = 37.85`

Or, as in the example from before, use the NumPy to calculate the standard deviation:

> அல்லது, முன் உதாரணத்தைப் போலவே, நிலையான விலகலைக் கணக்கிட NumPy ஐப் பயன்படுத்தவும்:

```python
# Use the NumPy std() method to find the standard deviation:

import numpy

speed = [32,111,138,28,59,77,97]

x = numpy.std(speed)

print(x)
```

## What are Percentiles?
## சதவீதங்கள் என்றால் என்ன?

Percentiles are used in statistics to give you a number that describes the value that a given percent of the values are lower than.

> கொடுக்கப்பட்ட சதவீத மதிப்புகள் குறைவாக இருக்கும் மதிப்பை விவரிக்கும் எண்ணை உங்களுக்கு வழங்க புள்ளிவிவரங்களில் சதவீதங்கள் பயன்படுத்தப்படுகின்றன.

Example: Let's say we have an array of the ages of all the people that live in a street.

> உதாரணம்: ஒரு தெருவில் வசிக்கும் அனைத்து மக்களின் வயதுகளின் வரிசை எங்களிடம் உள்ளது என்று வைத்துக்கொள்வோம்.

## `ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]`

What is the 75. percentile? The answer is 43, meaning that 75% of the people are 43 or younger.

> 75. சதவீதம் என்றால் என்ன? பதில் 43, அதாவது 75% மக்கள் 43 அல்லது அதற்கு குறைவானவர்கள்.

The NumPy module has a method for finding the specified percentile:

> 75. சதவீதம் என்றால் என்ன? பதில் 43, அதாவது 75% மக்கள் 43 அல்லது அதற்கு குறைவானவர்கள்.

```python
# Use the NumPy percentile() method to find the percentiles:

import numpy

ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]

x = numpy.percentile(ages, 75)

print(x)
```

## Normal Data Distribution
## சாதாரண தரவு விநியோகம்

In the previous chapter we learned how to create a completely random array, of a given size, and between two given values.

> முந்தைய அத்தியாயத்தில், கொடுக்கப்பட்ட அளவு மற்றும் கொடுக்கப்பட்ட இரண்டு மதிப்புகளுக்கு இடையில் முற்றிலும் சீரற்ற வரிசையை எவ்வாறு உருவாக்குவது என்பதைக் கற்றுக்கொண்டோம்.

In this chapter we will learn how to create an array where the values are concentrated around a given value.

> கொடுக்கப்பட்ட மதிப்பைச் சுற்றி மதிப்புகள் குவிந்திருக்கும் ஒரு வரிசையை எவ்வாறு உருவாக்குவது என்பதை இந்த அத்தியாயத்தில் கற்றுக்கொள்வோம்.

In probability theory this kind of data distribution is known as the *normal data distribution, or the Gaussian data distribution*, after the mathematician Carl Friedrich Gauss who came up with the formula of this data distribution.

> நிகழ்தகவு கோட்பாட்டில், இந்த வகையான தரவு விநியோகம் *சாதாரண தரவு விநியோகம் அல்லது காஸியன் தரவு விநியோகம்* என அறியப்படுகிறது, இந்த தரவு விநியோகத்தின் சூத்திரத்தை கொண்டு வந்த கணிதவியலாளர் கார்ல் ஃப்ரீட்ரிக் காஸ் என்பவருக்குப் பிறகு.

```python
# A typical normal data distribution:

import numpy
import matplotlib.pyplot as plt

x = numpy.random.normal(5.0, 1.0, 100000)

plt.hist(x, 100)
plt.show()
```

## How Does it Work?
## இது எப்படி வேலை செய்கிறது?

Python has methods for finding a relationship between data-points and to draw a line of linear regression. We will show you how to use these methods instead of going through the mathematic formula.

> தரவுப் புள்ளிகளுக்கு இடையேயான தொடர்பைக் கண்டறிவதற்கும் நேரியல் பின்னடைவுக் கோட்டை வரைவதற்கும் பைதான் முறைகளைக் கொண்டுள்ளது. கணித சூத்திரத்திற்குப் பதிலாக இந்த முறைகளை எவ்வாறு பயன்படுத்துவது என்பதை நாங்கள் உங்களுக்குக் காண்பிப்போம்.

In the example below, the x-axis represents age, and the y-axis represents speed. We have registered the age and speed of 13 cars as they were passing a tollbooth. Let us see if the data we collected could be used in a linear regression:

> கீழே உள்ள எடுத்துக்காட்டில், x-அச்சு வயதைக் குறிக்கிறது, மற்றும் y-அச்சு வேகத்தைக் குறிக்கிறது. 13 கார்கள் சுங்கச்சாவடியைக் கடந்து சென்றதால், அவற்றின் வயது மற்றும் வேகத்தைப் பதிவு செய்துள்ளோம். நாங்கள் சேகரித்த தரவு நேரியல் பின்னடைவில் பயன்படுத்தப்படுமா என்பதைப் பார்ப்போம்:

```python
# Start by drawing a scatter plot:

import matplotlib.pyplot as plt

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(x, y)
plt.show()
```

```python
# Import scipy and draw the line of Linear Regression:

import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

## R for Relationship
## உறவுக்கான R

It is important to know how the relationship between the values of the x-axis and the values of the y-axis is, if there are no relationship the linear regression can not be used to predict anything.

> x-அச்சின் மதிப்புகளுக்கும் y-அச்சின் மதிப்புகளுக்கும் இடையிலான உறவு எப்படி இருக்கிறது என்பதை அறிவது முக்கியம், எந்த உறவும் இல்லை என்றால் நேரியல் பின்னடைவு எதையும் கணிக்க பயன்படுத்த முடியாது.

This relationship - the coefficient of correlation - is called `r.`

> இந்த உறவு - தொடர்பு குணகம் - `r.` என்று அழைக்கப்படுகிறது

The `r` value ranges from -1 to 1, where 0 means no relationship, and 1 (and -1) means 100% related.

> `r` மதிப்பு -1 முதல் 1 வரை இருக்கும், இதில் 0 என்றால் உறவு இல்லை, 1 (மற்றும் -1) என்பது 100% தொடர்புடையது.

Python and the Scipy module will compute this value for you, all you have to do is feed it with the x and y values.

> Python மற்றும் Scipy தொகுதி உங்களுக்காக இந்த மதிப்பைக் கணக்கிடும், நீங்கள் செய்ய வேண்டியது x மற்றும் y மதிப்புகளுடன் அதை ஊட்ட வேண்டும்.

```python
# How well does my data fit in a linear regression?

from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

print(r)
```

### ` **Note:** The result -0.76 shows that there is a relationship, not perfect, but it indicates that we could use linear regression in future predictions.`

### ` **குறிப்பு:** முடிவு -0.76 ஒரு உறவு இருப்பதைக் காட்டுகிறது, சரியானதாக இல்லை, ஆனால் எதிர்கால கணிப்புகளில் நாம் நேரியல் பின்னடைவைப் பயன்படுத்தலாம் என்பதைக் குறிக்கிறது.`

## Predict Future Values
## எதிர்கால மதிப்புகளை கணிக்கவும்

Now we can use the information we have gathered to predict future values.

> இப்போது நாம் சேகரித்த தகவலை எதிர்கால மதிப்புகளை கணிக்க பயன்படுத்தலாம்.

Example: Let us try to predict the speed
of a 10 years old car.

> உதாரணம்: 10 வருட பழைய காரின் வேகத்தை கணிக்க முயற்சிப்போம்.

To do so, we need the same `myfunc()` function from the example above:

> அவ்வாறு செய்ய, மேலே உள்ள எடுத்துக்காட்டில் இருந்து அதே `myfunc()` செயல்பாடு நமக்குத் தேவை:

### `def myfunc(x):`
### `return slope * x + intercept`

```python
# Predict the speed of a 10 years old car:

from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

speed = myfunc(10)

print(speed)
```

```python
# These values for the x- and y-axis should result in a very bad fit for linear regression:

import matplotlib.pyplot as plt
from scipy import stats

x = [89,43,36,36,95,10,66,34,38,20,26,29,48,64,6,5,36,66,72,40]
y = [21,46,3,35,67,95,53,72,58,10,26,34,90,33,38,20,56,2,47,15]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

## Polynomial Regression
If your data points clearly will not fit a linear regression (a straight line through all data points), it might be ideal for polynomial regression.

> உங்கள் தரவுப் புள்ளிகள் ஒரு நேரியல் பின்னடைவுக்கு (எல்லா தரவுப் புள்ளிகளிலும் ஒரு நேர்கோடு) பொருந்தவில்லை என்றால், அது பல்லுறுப்புக்கோவை பின்னடைவுக்கு ஏற்றதாக இருக்கலாம்.

Polynomial regression, like linear regression, uses the relationship between the variables x and y to find the best way to draw a line through the data points.

> பல்லுறுப்புக்கோவை பின்னடைவு, நேரியல் பின்னடைவு போன்றது, தரவு புள்ளிகள் வழியாக ஒரு கோட்டை வரைய சிறந்த வழியைக் கண்டறிய x மற்றும் y மாறிகளுக்கு இடையிலான உறவைப் பயன்படுத்துகிறது.

## R-Squared
## R-சதுரம்

It is important to know how well the relationship between the values of the x- and y-axis is, if there are no relationship the polynomial regression can not be used to predict anything.
> x- மற்றும் y- அச்சின் மதிப்புகளுக்கு இடையே உள்ள உறவு எவ்வளவு நன்றாக உள்ளது என்பதை அறிவது முக்கியம், எந்த உறவும் இல்லை என்றால் பல்லுறுப்புக்கோவை பின்னடைவை எதையும் கணிக்க பயன்படுத்த முடியாது.

The relationship is measured with a value called the r-squared.

> உறவுமுறை r-squared எனப்படும் மதிப்பைக் கொண்டு அளவிடப்படுகிறது.

The r-squared value ranges from 0 to 1, where 0 means no relationship, and 1 means 100% related.

> r-சதுர மதிப்பு 0 முதல் 1 வரை இருக்கும், இதில் 0 என்றால் உறவு இல்லை, 1 என்றால் 100% தொடர்புடையது.

Python and the Sklearn module will compute this value for you, all you have to do is feed it with the x and y arrays:

> Python மற்றும் Sklearn தொகுதி உங்களுக்காக இந்த மதிப்பைக் கணக்கிடும், நீங்கள் செய்ய வேண்டியது x மற்றும் y வரிசைகள் மூலம் அதை ஊட்டுவதுதான்:

```python
# How well does my data fit in a polynomial regression?

import numpy
from sklearn.metrics import r2_score

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

print(r2_score(y, mymodel(x)))
```

### `Note: The result 0.94 shows that there is a very good relationship, and we can use polynomial regression in future predictions.`

> ### `குறிப்பு: முடிவு 0.94 ஒரு நல்ல உறவு இருப்பதைக் காட்டுகிறது, மேலும் எதிர்கால கணிப்புகளில் நாம் பல்லுறுப்புக்கோவை பின்னடைவைப் பயன்படுத்தலாம்.`

This object has a method called `fit()` that takes the independent and dependent values as parameters and fills the regression object with data that describes the relationship:

## Coefficient
## குணகம்
The coefficient is a factor that describes the relationship with an unknown variable.

> குணகம் என்பது அறியப்படாத மாறியுடனான உறவை விவரிக்கும் ஒரு காரணியாகும்.

Example: if `x` is a variable, then `2x` is `x` two times. `x` is the unknown variable, and the number `2` is the coefficient.

> எடுத்துக்காட்டு: `x` என்பது மாறி என்றால், `2x` என்பது `x` என்பது இரண்டு முறை. `x` என்பது தெரியாத மாறி, மேலும் `2` எண் குணகம்.

In this case, we can ask for the coefficient value of weight against CO2, and for volume against CO2. The answer(s) we get tells us what would happen if we increase, or decrease, one of the independent values.

> இந்த வழக்கில், CO2 க்கு எதிரான எடையின் குணக மதிப்பையும், CO2 க்கு எதிரான அளவையும் கேட்கலாம். நாம் பெறும் பதில்(கள்) நாம் சுதந்திரமான மதிப்புகளில் ஒன்றைக் கூட்டினால் அல்லது குறைத்தால் என்ன நடக்கும் என்று நமக்குச் சொல்கிறது.

```python
# Print the coefficient values of the regression object:

import pandas
from sklearn import linear_model

df = pandas.read_csv("data.csv")

X = df[['Weight', 'Volume']]
y = df['CO2']

regr = linear_model.LinearRegression()
regr.fit(X, y)

print(regr.coef_)
```

## Scale Features

## அளவு அம்சங்கள்

When your data has different values, and even different measurement units, it can be difficult to compare them. What is kilograms compared to meters? Or altitude compared to time?

> உங்கள் தரவு வெவ்வேறு மதிப்புகள் மற்றும் வெவ்வேறு அளவீட்டு அலகுகளைக் கொண்டிருக்கும் போது, ​​அவற்றை ஒப்பிடுவது கடினமாக இருக்கும். மீட்டருடன் ஒப்பிடும்போது கிலோகிராம் என்ன? அல்லது நேரத்துடன் ஒப்பிடும்போது உயரமா?

There are different methods for scaling data, in this tutorial we will use a method called standardization.

> தரவை அளவிடுவதற்கு வெவ்வேறு முறைகள் உள்ளன, இந்த டுடோரியலில் தரநிலைப்படுத்தல் என்ற முறையைப் பயன்படுத்துவோம்.

The standardization method uses this formula:

### `z = (x - u) / s`

Where `z` is the new value, `x` is the original value, `u` is the mean and `s` is the standard deviation.

> `z` என்பது புதிய மதிப்பு, `x` என்பது அசல் மதிப்பு, `u` என்பது சராசரி மற்றும் `s` என்பது நிலையான விலகல்.

If you take the **weight** column from the data set above, the first value is 790, and the scaled value will be:

> மேலே உள்ள தரவுத் தொகுப்பிலிருந்து **எடை** நெடுவரிசையை நீங்கள் எடுத்தால், முதல் மதிப்பு 790 மற்றும் அளவிடப்பட்ட மதிப்பு:

### `(790 - 1292.23) / 238.74 = -2.1`

If you take the **volume** column from the data set above, the first value is 1.0, and the scaled value will be:

> மேலே உள்ள தரவுத் தொகுப்பிலிருந்து **தொகுதி** நெடுவரிசையை நீங்கள் எடுத்தால், முதல் மதிப்பு 1.0 மற்றும் அளவிடப்பட்ட மதிப்பு:

### `(1.0 - 1.61) / 0.38 = -1.59`

Now you can compare -2.1 with -1.59 instead of comparing 790 with 1.0.

> இப்போது நீங்கள் 790 ஐ 1.0 உடன் ஒப்பிடுவதற்குப் பதிலாக -2.1 ஐ -1.59 உடன் ஒப்பிடலாம்.

You do not have to do this manually, the Python sklearn module has a method called `StandardScaler()` which returns a Scaler object with methods for transforming data sets.

> இதை நீங்கள் கைமுறையாகச் செய்ய வேண்டியதில்லை, பைதான் ஸ்க்லேர்ன் தொகுதிக்கு `ஸ்டாண்டர்ட் ஸ்கேலர்()` எனப்படும் ஒரு முறை உள்ளது, இது தரவுத் தொகுப்புகளை மாற்றும் முறைகளுடன் ஸ்கேலர் பொருளை வழங்குகிறது.


```python
# Scale all values in the Weight and Volume columns:

import pandas
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

df = pandas.read_csv("data.csv")

X = df[['Weight', 'Volume']]

scaledX = scale.fit_transform(X)

print(scaledX)
```

Note that the first two values are -2.1 and -1.59, which corresponds to our calculations:

## Predict CO2 Values
## CO2 மதிப்புகளைக் கணிக்கவும்

The task in the *Multiple Regression chapter* was to predict the CO2 emission from a car when you only knew its weight and volume.

> *மல்டிபிள் ரிக்ரஷன் அத்தியாயத்தில்* உள்ள பணியானது, ஒரு காரின் எடை மற்றும் கன அளவு மட்டும் தெரிந்தால், அதில் இருந்து CO2 உமிழ்வைக் கணிப்பது.


When the data set is scaled, you will have to use the scale when you predict values:

> தரவுத் தொகுப்பு அளவிடப்படும்போது, ​​நீங்கள் மதிப்புகளைக் கணிக்கும்போது அளவைப் பயன்படுத்த வேண்டும்:


```python
# Predict the CO2 emission from a 1.3 liter car that weighs 2300 kilograms:

import pandas
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

df = pandas.read_csv("data.csv")

X = df[['Weight', 'Volume']]
y = df['CO2']

scaledX = scale.fit_transform(X)

regr = linear_model.LinearRegression()
regr.fit(scaledX, y)

scaled = scale.transform([[2300, 1.3]])

predictedCO2 = regr.predict([scaled[0]])
print(predictedCO2)
```

# உங்கள் மாதிரியை மதிப்பிடுங்கள்

In Machine Learning we create models to predict the outcome of certain events, like in the previous chapter where we predicted the CO2 emission of a car when we knew the weight and engine size.

> இயந்திரக் கற்றலில், சில நிகழ்வுகளின் விளைவுகளைக் கணிக்க மாதிரிகளை உருவாக்குகிறோம், முந்தைய அத்தியாயத்தில், எடை மற்றும் எஞ்சின் அளவை அறிந்தால் காரின் CO2 உமிழ்வைக் கணித்ததைப் போல.

To measure if the model is good enough, we can use a method called Train/Test.

> மாதிரி போதுமானதாக இருக்கிறதா என்பதை அளவிட, ரயில்/சோதனை என்ற முறையைப் பயன்படுத்தலாம்.

## What is Train/Test
## ரயில்/சோதனை என்றால் என்ன

Train/Test is a method to measure the accuracy of your model.

> ரயில்/சோதனை என்பது உங்கள் மாதிரியின் துல்லியத்தை அளவிடுவதற்கான ஒரு முறையாகும்.

It is called Train/Test because you split the data set into two sets: a training set and a testing set.

> நீங்கள் தரவை இரண்டு செட்களாகப் பிரிப்பதால் இது ரயில்/சோதனை என்று அழைக்கப்படுகிறது: பயிற்சித் தொகுப்பு மற்றும் சோதனைத் தொகுப்பு.

## `80% for training, and 20% for testing.`
You *train* the model using the training set.

You *test* the model using the testing set.

## `Train the model means create the model.`

## `Test the model means test the accuracy of the model.`

## Start With a Data Set
## தரவு தொகுப்புடன் தொடங்கவும்

Start with a data set you want to test.

> நீங்கள் சோதிக்க விரும்பும் தரவு தொகுப்புடன் தொடங்கவும்.

Our data set illustrates 100 customers in a shop, and their shopping habits.

> எங்கள் தரவுத் தொகுப்பு ஒரு கடையில் உள்ள 100 வாடிக்கையாளர்களையும் அவர்களின் ஷாப்பிங் பழக்கத்தையும் விளக்குகிறது.

```python
import numpy
import matplotlib.pyplot as plt
numpy.random.seed(2)

x = numpy.random.normal(3, 1, 100)
y = numpy.random.normal(150, 40, 100) / x

plt.scatter(x, y)
plt.show()
```

### விளைவாக:

> x அச்சு என்பது வாங்குவதற்கு முன் எத்தனை நிமிடங்களைக் குறிக்கிறது.

The y axis represents the amount of money spent on the purchase.

> y அச்சு என்பது வாங்குவதற்கு செலவழிக்கப்பட்ட பணத்தின் அளவைக் குறிக்கிறது.

The training set should be a random selection of 80% of the original data.

> பயிற்சித் தொகுப்பானது 80% அசல் தரவின் சீரற்ற தேர்வாக இருக்க வேண்டும்.

The testing set should be the remaining 20%.

> சோதனைத் தொகுப்பு மீதமுள்ள 20% ஆக இருக்க வேண்டும்.

> சோதனைத் தொகுப்பு முற்றிலும் வேறுபட்டதாக இல்லை என்பதை உறுதிப்படுத்த, சோதனைத் தொகுப்பையும் பார்ப்போம்.

### Fit the Data Set
### தரவுத் தொகுப்பைப் பொருத்தவும்

What does the data set look like? In my opinion I think the best fit would be a polynomial regression, so let us draw a line of polynomial regression.

> தரவுத் தொகுப்பு எப்படி இருக்கும்? என் கருத்துப்படி, பல்லுறுப்புக்கோவை பின்னடைவு மிகவும் பொருத்தமானது என்று நான் நினைக்கிறேன், எனவே பல்லுறுப்புக்கோவை பின்னடைவின் கோட்டை வரைவோம்.

To draw a line through the data points, we use the `plot()` method of the matplotlib module:

> தரவுப் புள்ளிகள் மூலம் ஒரு கோடு வரைவதற்கு, matplotlib தொகுதியின் `plot()` முறையைப் பயன்படுத்துகிறோம்:

```python
# Draw a polynomial regression line through the data points:

import numpy
import matplotlib.pyplot as plt
numpy.random.seed(2)

x = numpy.random.normal(3, 1, 100)
y = numpy.random.normal(150, 40, 100) / x

train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]

mymodel = numpy.poly1d(numpy.polyfit(train_x, train_y, 4))

myline = numpy.linspace(0, 6, 100)

plt.scatter(train_x, train_y)
plt.plot(myline, mymodel(myline))
plt.show()
```

The result can back my suggestion of the data set fitting a polynomial regression, even though it would give us some weird results if we try to predict values outside of the data set. Example: the line indicates that a customer spending 6 minutes in the shop would make a purchase worth 200. That is probably a sign of overfitting.

> தரவுத் தொகுப்பிற்கு வெளியே உள்ள மதிப்புகளைக் கணிக்க முயற்சித்தால், அது நமக்கு சில வித்தியாசமான முடிவுகளைத் தரும். உதாரணம்: கடையில் 6 நிமிடங்கள் செலவழிக்கும் வாடிக்கையாளர் 200 மதிப்புள்ள கொள்முதல் செய்வார் என்பதை வரி குறிக்கிறது.

But what about the R-squared score? The R-squared score is a good indicator of how well my data set is fitting the model.

> ஆனால் R ஸ்கொயர் ஸ்கோர் பற்றி என்ன? R-ஸ்கொயர் ஸ்கோர் என்பது எனது தரவுத் தொகுப்பு மாதிரிக்கு எவ்வளவு நன்றாகப் பொருந்துகிறது என்பதற்கான நல்ல குறிகாட்டியாகும்


## R2
Remember R2, also known as R-squared?
> R-squared என்றும் அழைக்கப்படும் R2 நினைவிருக்கிறதா?

It measures the relationship between the x axis and the y axis, and the value ranges from 0 to 1, where 0 means no relationship, and 1 means totally related.

> இது x அச்சுக்கும் y அச்சுக்கும் இடையிலான உறவை அளவிடுகிறது, மேலும் மதிப்பு 0 முதல் 1 வரை இருக்கும், இதில் 0 என்றால் உறவு இல்லை, 1 என்பது முற்றிலும் தொடர்புடையது.

The sklearn module has a method called `r2_score()` that will help us find this relationship.

> sklearn தொகுதியில் `r2_score()` என்ற முறை உள்ளது, இது இந்த உறவைக் கண்டறிய உதவும்.

In this case we would like to measure the relationship between the minutes a customer stays in the shop and how much money they spend.

> இந்த விஷயத்தில் வாடிக்கையாளர் கடையில் தங்கியிருக்கும் நிமிடங்களுக்கும் அவர்கள் எவ்வளவு பணம் செலவழிக்கிறார்கள் என்பதற்கும் இடையிலான உறவை அளவிட விரும்புகிறோம்.

```python
# How well does my training data fit in a polynomial regression?

import numpy
from sklearn.metrics import r2_score
numpy.random.seed(2)

x = numpy.random.normal(3, 1, 100)
y = numpy.random.normal(150, 40, 100) / x

train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]

mymodel = numpy.poly1d(numpy.polyfit(train_x, train_y, 4))

r2 = r2_score(train_y, mymodel(train_x))

print(r2)
```

### `Note: The result 0.799 shows that there is a OK relationship.`

### `குறிப்பு: முடிவு 0.799 சரி உறவு இருப்பதைக் காட்டுகிறது.`


## Bring in the Testing Set
## சோதனை தொகுப்பை கொண்டு வாருங்கள்

Now we have made a model that is OK, at least when it comes to training data.

>இப்போது நாங்கள் ஒரு மாதிரியை உருவாக்கியுள்ளோம், குறைந்தபட்சம் பயிற்சி தரவு வரும்போது சரி.

Now we want to test the model with the testing data as well, to see if gives us the same result.

> இப்போது மாதிரியை சோதனைத் தரவுகளுடன் சோதிக்க விரும்புகிறோம், அதே முடிவை நமக்குத் தருகிறதா என்பதைப் பார்க்கவும்.

```python
# Let us find the R2 score when using testing data:

import numpy
from sklearn.metrics import r2_score
numpy.random.seed(2)

x = numpy.random.normal(3, 1, 100)
y = numpy.random.normal(150, 40, 100) / x

train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]

mymodel = numpy.poly1d(numpy.polyfit(train_x, train_y, 4))

r2 = r2_score(test_y, mymodel(test_x))

print(r2)
```

### `Note: The result 0.809 shows that the model fits the testing set as well, and we are confident that we can use the model to predict future values.`

> ### `குறிப்பு: முடிவு 0.809 மாதிரியானது சோதனைத் தொகுப்பிற்கும் பொருந்துகிறது என்பதைக் காட்டுகிறது, மேலும் எதிர்கால மதிப்புகளைக் கணிக்க மாதிரியைப் பயன்படுத்தலாம் என்பதில் நாங்கள் உறுதியாக உள்ளோம்.`

## மதிப்புகளை கணிக்கவும்

## How Does it Work?
First, read the dataset with pandas:

```python
# Read and print the data set:

import pandas

df = pandas.read_csv("data.csv")

print(df)
```

To make a decision tree, all data has to be numerical.
> முடிவெடுக்கும் மரத்தை உருவாக்க, எல்லா தரவுகளும் எண்ணாக இருக்க வேண்டும்.

We have to convert the non numerical columns 'Nationality' and 'Go' into numerical values.

> எண் அல்லாத நெடுவரிசைகளான 'தேசியம்' மற்றும் 'கோ' ஆகியவற்றை எண் மதிப்புகளாக மாற்ற வேண்டும்.

Pandas has a `map()` method that takes a dictionary with information on how to convert the values.

> மதிப்புகளை எவ்வாறு மாற்றுவது என்பது குறித்த தகவல்களைக் கொண்ட அகராதியை எடுக்கும் `வரைபடம்()` முறையை பாண்டாஸ் கொண்டுள்ளது.

    {'UK': 0, 'USA': 1, 'N': 2}

Means convert the values 'UK' to 0, 'USA' to 1, and 'N' to 2.

> அதாவது 'UK' மதிப்புகளை 0 ஆகவும், 'USA' ஐ 1 ஆகவும், 'N' ஐ 2 ஆகவும் மாற்றும்.

```python
# Change string values into numerical values:

d = {'UK': 0, 'USA': 1, 'N': 2}
df['Nationality'] = df['Nationality'].map(d)
d = {'YES': 1, 'NO': 0}
df['Go'] = df['Go'].map(d)

print(df)
```

Then we have to separate the feature columns from the target column.

> பின்னர் இலக்கு நெடுவரிசையிலிருந்து அம்ச நெடுவரிசைகளை பிரிக்க வேண்டும்.

The feature columns are the columns that we try to predict from, and the target column is the column with the values we try to predict.

> அம்ச நெடுவரிசைகள் என்பது நாம் கணிக்க முயற்சிக்கும் நெடுவரிசைகள் மற்றும் இலக்கு நெடுவரிசை என்பது நாம் கணிக்க முயற்சிக்கும் மதிப்புகளைக் கொண்ட நெடுவரிசையாகும்.

```python
# X is the feature columns, y is the target column:

features = ['Age', 'Experience', 'Rank', 'Nationality']

X = df[features]
y = df['Go']

print(X)
print(y)
```

Now we can create the actual decision tree, fit it with our details. Start by importing the modules we need:

> இப்போது நாம் உண்மையான முடிவு மரத்தை உருவாக்கலாம், அதை எங்கள் விவரங்களுடன் பொருத்தலாம். நமக்குத் தேவையான தொகுதிகளை இறக்குமதி செய்வதன் மூலம் தொடங்கவும்:

```python
# Create and display a Decision Tree:

import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

df = pandas.read_csv("data.csv")

d = {'UK': 0, 'USA': 1, 'N': 2}
df['Nationality'] = df['Nationality'].map(d)
d = {'YES': 1, 'NO': 0}
df['Go'] = df['Go'].map(d)

features = ['Age', 'Experience', 'Rank', 'Nationality']

X = df[features]
y = df['Go']

dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

tree.plot_tree(dtree, feature_names=features)
```

    Gini
    There are many ways to split the samples, we use the GINI method in this tutorial.

    The Gini method uses this formula:

    Gini = 1 - (x/n)2 - (y/n)2

    Where x is the number of positive answers("GO"), n is the number of samples, and y is the number of negative answers ("NO"), which gives us this calculation:

    1 - (7 / 13)2 - (6 / 13)2 = 0.497

    கினி
    மாதிரிகளைப் பிரிக்க பல வழிகள் உள்ளன, இந்த டுடோரியலில் GINI முறையைப் பயன்படுத்துகிறோம்.

    கினி முறை இந்த சூத்திரத்தைப் பயன்படுத்துகிறது:

    கினி = 1 - (x/n)2 - (y/n)2

    x என்பது நேர்மறை பதில்களின் எண்ணிக்கை ("GO"), n என்பது மாதிரிகளின் எண்ணிக்கை, மற்றும் y என்பது எதிர்மறை பதில்களின் எண்ணிக்கை ("NO"), இது இந்தக் கணக்கீட்டை நமக்கு வழங்குகிறது:

`value = [1, 0]` means that 1 will get a "NO" and 0 will get a "GO".

> `மதிப்பு = [1, 0]` என்பது 1 க்கு "NO" கிடைக்கும் மற்றும் 0 க்கு "GO" கிடைக்கும்.


## Predict Values
## மதிப்புகளை கணிக்கவும்

We can use the Decision Tree to predict new values.

> புதிய மதிப்புகளைக் கணிக்க முடிவு மரத்தைப் பயன்படுத்தலாம்.

Example: Should I go see a show starring a 40 years old American comedian, with 10 years of experience, and a comedy ranking of 7?

> உதாரணம்: 10 வருட அனுபவமும், 7வது நகைச்சுவை தரவரிசையும் கொண்ட, 40 வயது அமெரிக்க நகைச்சுவை நடிகரின் ஒரு நிகழ்ச்சியைப் பார்க்க நான் செல்ல வேண்டுமா?

```python
# Use predict() method to predict new values:

import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier

df = pandas.read_csv("data.csv")

d = {'UK': 0, 'USA': 1, 'N': 2}
df['Nationality'] = df['Nationality'].map(d)
d = {'YES': 1, 'NO': 0}
df['Go'] = df['Go'].map(d)

features = ['Age', 'Experience', 'Rank', 'Nationality']
X = df[features]
y = df['Go']

dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

print(dtree.predict([[40, 10, 7, 1]]))

print("[1] means 'GO'")
print("[0] means 'NO'")
```

```python
# What would the answer be if the comedy rank was 6?

import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier

df = pandas.read_csv("data.csv")

d = {'UK': 0, 'USA': 1, 'N': 2}
df['Nationality'] = df['Nationality'].map(d)
d = {'YES': 1, 'NO': 0}
df['Go'] = df['Go'].map(d)

features = ['Age', 'Experience', 'Rank', 'Nationality']
X = df[features]
y = df['Go']

dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

print(dtree.predict([[40, 10, 6, 1]]))

print("[1] means 'GO'")
print("[0] means 'NO'")
```

## Different Results
## வெவ்வேறு முடிவுகள்

You will see that the Decision Tree gives you different results if you run it enough times, even if you feed it with the same data.

> டெசிஷன் ட்ரீயை நீங்கள் ஒரே டேட்டாவுடன் ஊட்டினாலும், போதுமான முறை இயக்கினால் வெவ்வேறு முடிவுகளைத் தருவதை நீங்கள் காண்பீர்கள்.

That is because the Decision Tree does not give us a 100% certain answer. It is based on the probability of an outcome, and the answer will vary.

> ஏனென்றால், முடிவு மரம் நமக்கு 100% உறுதியான பதிலைத் தரவில்லை. இது ஒரு முடிவின் நிகழ்தகவை அடிப்படையாகக் கொண்டது, மேலும் பதில் மாறுபடும்.
